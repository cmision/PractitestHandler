```python
import pandas as pd
import numpy as np
```

```python
frameA = pd.read_excel('A.xlsx', sheet_name='Data',index_col='ID')
frameB = pd.read_excel('B.xlsx', sheet_name='Data',index_col='ID')
frameC = pd.read_excel('C.xlsx', sheet_name='Data',index_col='ID')
```

```python
frameA.head()
```

```python
frameA.rename(columns = {'Linked Tests':'B'}, inplace = True) 
frameA['B'] = frameA['B'].str.replace("\[\[", "[")
frameA['B'] = frameA['B'].str.replace("\]\]", "]")
frameA['B'] = frameA['B'].str.replace(" \"", "|")
frameA['B'] = frameA['B'].str.replace("\"\] \[", ",")
frameA['B'] = frameA.B.apply(lambda x: x[1:-1].split(','))
frameA=frameA.explode('B')
```

```python
frameA.head()
```

```python
frameA[['TestIndex','TestName']] = frameA['B'].str.split('|',expand=True)
frameA['TestIndex']=frameA['TestIndex'].replace('',np.NaN)
```
